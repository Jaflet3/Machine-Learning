```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("scatter plot of total bill vs. tip")
plt.xlabel("total bill()")
plt.ylabel("tip($)")
plt.show()
```

```python
sns.barplot(x="total_bill",y="tip",data=tips)
plt.title("scatter plot of total bill vs. tip")
plt.xlabel("total bill()")
plt.ylabel("tip($)")
plt.show()
```

```python
sns.boxplot(x="total_bill",y="tip",data=tips)
plt.title("scatter plot of total bill vs. tip")
plt.xlabel("total bill()")
plt.ylabel("tip($)")
plt.show()
```

```python
sns.violinplot(x="total_bill",y="tip",data=tips)
plt.title("scatter plot of total bill vs. tip")
plt.xlabel("total bill()")
plt.ylabel("tip($)")
plt.show()
```

```python
sns.lineplotplot(x="total_bill",y="tip",data=tips)
plt.title("scatter plot of total bill vs. tip")
plt.xlabel("total bill()")
plt.ylabel("tip($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris")
correlation_matrix=iris.select_dtypes(include=['number']).corr()
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm")
plt.title("correlation matrix of iris dataset")
plt.show()
```

```python
sns.pairplot(iris)
plt.title("pair plot")
plt.show()

```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex")
plt.title("joint plot")
plt.show()
```
